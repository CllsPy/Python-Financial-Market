```python
import pandas as pd #importar e manipular dados
import yfinance as yf #pegar dados de um site: Web Scratch
import matplotlib.pyplot as plt #gráficos
import numpy as np #manipular arrays/matrizes
```

```python
simbolo = 'PETR4.SA'
data_inicio = '2020-01-01'
data_fim = '2023-01-01'

data = yf.download(simbolo, start=data_inicio, end=data_fim)
data.head()

```

```python
data['Resultado'] = (data['Adj Close'] / data['Adj Close'].shift(1)) -1
data.head()
```

```python
for column in data.columns:

    data[column].plot(color = 'orange')
    plt.xlabel('Ano')
    plt.ylabel(column)
    plt.title(f'Gráfico de {column}')
    plt.legend()
    plt.show()
    print("\n")
```

```python
media_simples_resultados = data [ 'Resultado' ].mean ()
print(media_simples_resultados)
```

```python
media_anual_Aj_Close = data[ 'Resultado'].mean () * 250
print(f'Portanto o Redimento da {simbolo} entre {data_inicio} e {data_fim} foi de:', (str(round( media_anual_Aj_Close, 3 ) * 100 ) + '%'))
```
